```python
import pandas as pd 
import numpy as np
```

```python
WIKI_URL = "https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=890001695"
dfs = pd.read_html(WIKI_URL, header=0)
df=dfs[0]
```

```python
df1 = df[df['Borough']!='Not assigned']
```

```python
df1.Neighbourhood[df1.Neighbourhood == 'Not assigned'] = "Queen's Park"
```

```python
df_postcode = df1.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
```
